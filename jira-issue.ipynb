```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import sqlite3
conn = sqlite3.connect('sqlite-database.db')
```

```python
issues = pd.read_sql_query("SELECT * from jira_issue", conn, parse_dates=["created","updated","resolved"])
```
